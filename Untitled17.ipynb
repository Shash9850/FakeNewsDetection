```python
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
true_news = pd.read_csv('True.csv')
fake_news = pd.read_csv('Fake.csv')

true_news.head(10)
```

```python
fake_news.head(10)
```

```python
true_news['label']=1
true_news.head(10)
```

```python
fake_news['label']=0


fake_news.head(10)
```

```python

# Ensure data types
true_news['label'] = true_news['label'].astype(int)
fake_news['label'] = fake_news['label'].astype(int)

# Concatenate DataFrames
news = pd.concat([true_news, fake_news], axis=0).reset_index(drop=True)

# Verify result
print(news.head())
```

```python
news=news.drop(['title','subject','date'],axis=1)
news.head(10)
```

```python
news = news.sample(frac=1)
news.head(15)
```

```python
news.reset_index(inplace=True)
news.head(15)
```

```python
news.columns
```

```python
def wordopt(text):
    #convert to lowercase
    text = text.lower()
    #remove urls
    text = re.sub('https?://\S+|www\.\S+','',text)
    #remove html tags
    text = re.sub(r'<.*?>','',text)
    #remove puntuation
    text = re.sub(r'[^\w\s]','',text)
    #remove digits
    text = re.sub(r'\d','',text)
    #remove newline characters
    text = re.sub(r'\n','',text)
    return text

news['text'] = news['text'].apply(wordopt)
```

```python
x = news['text']

y=news['label']
```

```python
# Check train-test split and vectorizer
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Fit-transform and transform
vectorizer = TfidfVectorizer()
xv_train = vectorizer.fit_transform(x_train)
xv_test = vectorizer.transform(x_test)

# Debugging shapes
print("xv_train shape:", xv_train.shape)
print("xv_test shape:", xv_test.shape)

# Logistic Regression
LReg = LogisticRegression()
LReg.fit(xv_train, y_train)

# Predict
L_pred = LReg.predict(xv_test)
print("Predictions:", L_pred[:5])
```

```python
LReg.score(xv_test,y_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,L_pred))
```
